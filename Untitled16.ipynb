1. LOGISTIC REGRESSION

Will help us find the binomial answer (whether an employee stayed or left)

Assumptions for the model:

Outcome variable is categorical

Observations are independent of each other

No severe multicollinearity among X variables

No extreme outliers

Linear relationship between each X variable and the logit of the outcome variable

Sufficiently large sample size

```python
df_log = clean_df.copy()
```

```python
df_log['salary'] = df_log['salary'].astype('category').cat.set_categories(['low', 'medium', 'high']).cat.codes

# use encoding for departments
df_log = pd.get_dummies(df_log, drop_first=True, dtype='int')
```

```python
df_log.head(10)
```

Check for multicollinearity

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_log[['satisfaction_level', 'last_evaluation', 'number_project', 'average_monthly_hours', 'tenure']].corr() ,cmap='crest', annot=True)
```

```python
#remove the outliers we found before

df_log = df_log[(df_log['tenure']<=upper_limit) | (df_log['tenure']<lower_limit)]
```

```python
df_log.head(10)
```

```python
#divide into x and y variables

X = df_log.drop('left', axis=1)
y = df_log['left']
```

```python
#split into training and testing model

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.24, random_state=42, stratify=y)
```

```python
clf = LogisticRegression(random_state=42,max_iter=500).fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

EVALUATION


```python
#compute values for Confusion Matrix

cm= confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
```

Interpretation

The model was able to find out 2081 people who did not leave, who actually did not leave, 148 people who did not leave predicted as left.

334 people who left but actually did not leave,118 people who left and who actually left

```python
def read_pickle(path, saved_model_name:str):
    '''
    In:
        path:             path to folder where you want to read from
        saved_model_name: filename of pickled model you want to read in

    Out:
        model: the pickled model
    '''
    with open(path + saved_model_name + '.pickle', 'rb') as to_read:
        model = pickle.load(to_read)

    return model
```

It is reasonable to be cautious about the unusually high accuracy results, as this may indicate the presence of data leakage. Data leakage occurs when information from the training dataset unintentionally appears in the test dataset.

In our case, satisfaction_level and average_monthly_hours may be contributing to this issue. An employee with a low evaluation score may already be on the verge of leaving, which the model could be learning directly. Similarly, average_monthly_hours may decrease when an employee is planning to quit and becomes disengaged from their manager, making this information a strong but potentially leaked predictor.

In order to tackle this issue, we can perform Feature Engineering.

```python
#remove satisfaction_levels

df_log = df_log.drop('satisfaction_level', axis=1)

```
